```python
appartementData=pd.read_csv('https://raw.githubusercontent.com/ardhiraka/PFDS_sources/master/property_data.csv',
                           na_values=["na","--"])
appartementData
```

```python
allowedValue=["Y","N"]
idx=0
for row in appartementData["OWN_OCCUPIED"]:
    try:
        allowedValue.index(row)
    except ValueError:
        appartementData.loc[idx,"OWN_OCCUPIED"]=np.nan
    idx+=1
```

```python
appartementData
```

```python
idx=0
for row in appartementData['NUM_BATH']:
    try :
        int(row)
    except ValueError:
        appartementData.loc[idx, 'NUM_BATH']=np.nan
    idx+=1
```
