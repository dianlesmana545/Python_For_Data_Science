```python
movieDB.sort_values(by='star_rating', axis=0, ascending=False)
```

```python
import re
movieDB[movieDB.title.str.contains("^sp",flags=re.IGNORECASE)]
```

```python
grupMovie=movieDB.groupby(['genre'])['title']
for key, value in grupMovie.groups.items() :
    for valInx in value :
        print(movieDB[movieDB.genre=key])
```
