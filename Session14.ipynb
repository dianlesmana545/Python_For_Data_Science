```python
from sklearn.datasets import load_breast_cancer
```

```python
breast = load_breast_cancer()
breast
```

```python
breast_data = breast.data
breast_data
```

```python
breast_data.shape
```

```python
breast_label = breast.target
breast_label
```

```python
import numpy as np
```

```python
labels = np.reshape(breast_label,(569,1))
labels
```

```python
final_breast_data = np.concatenate([breast_data,labels], axis=1)
```

```python
final_breast_data.shape
```

```python
import pandas as pd
```

```python
breast_dataset = pd.DataFrame(final_breast_data)
breast_dataset
```

```python
features = breast.feature_names
```

```python
features
```

```python
features_labels = np.append(features,'label')
```

```python
breast_dataset.columns = features_labels
```

```python
breast_dataset.head()
```

```python
breast_dataset["label_name"] = np.where(breast_dataset["label"]==0, "Benign", "Maligant")
```

```python
breast_dataset["label_name"].tail(30)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
x = breast_dataset.loc[:,features].values
x = StandardScaler().fit_transform(x)
x
```

```python
x.shape
```

```python
column_name = ['feature'+str(i) for i in range(x.shape[1])]
```

```python
column_name
```

```python
normalize_breast_df = pd.DataFrame(x, columns=column_name)
```

```python
normalize_breast_df.head()
```

```python
from sklearn.decomposition import PCA
```

```python
pca_model = PCA(n_components = 2)
principal_com_breast = pca_model.fit_transform(x)
```

```python
principal_com_breast
```

```python
pca_model.explained_variance_ratio_
```

```python
normalize_breast_df = pd.DataFrame(principal_com_breast, columns=['principal component 1','principal component 2'])
```

```python
normalize_breast_df
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
plt.figure()
plt.figure(figsize=(15,15))
plt.xticks(fontsize=12)
plt.yticks(fontsize=14)
plt.xlabel("pricipal component -1")
plt.ylabel("pricipal component -2")
plt.title("Principal component breast cances")

targets = ["Benign", "Maligant"]
color = ["r","g"]
for target, color in zip(targets,color):
    idxToKeep = breast_dataset["label_name"]== target
    plt.scatter(normalize_breast_df.loc[idxToKeep,"principal component 1"],
                normalize_breast_df.loc[idxToKeep,"principal component 2"],
               c=color, s=40)
```
